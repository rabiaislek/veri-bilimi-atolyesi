```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
data = pd.read_csv("mental-illnesses-prevalence.csv") 
data
```

```python
data = pd.DataFrame(data) 
```

```python
depressive = data.groupby("Year")[["Depressive disorders (share of population) - Sex: Both - Age: Age-standardized"]].mean().reset_index()
depressive
```

```python
plt.figure(figsize=(12, 6))
sns.barplot(x='Year', y='Depressive disorders (share of population) - Sex: Both - Age: Age-standardized', data=depressive)
plt.title('Average Depressive Disorders Over the Years')
plt.xlabel('Year')
plt.ylabel('Share of Population (%)')
plt.xticks(rotation=45)
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
Depressive = data.groupby("Year")[["Depressive disorders (share of population) - Sex: Both - Age: Age-standardized"]].mean().mean().item()
Schizophrenia = data.groupby("Year")[["Schizophrenia disorders (share of population) - Sex: Both - Age: Age-standardized"]].mean().mean().item()
Anxiety = data.groupby("Year")[["Anxiety disorders (share of population) - Sex: Both - Age: Age-standardized"]].mean().mean().item()
Bipolar = data.groupby("Year")[["Bipolar disorders (share of population) - Sex: Both - Age: Age-standardized"]].mean().mean().item()
Eating = data.groupby("Year")[["Depressive disorders (share of population) - Sex: Both - Age: Age-standardized"]].mean().mean().item()

disorders = [Depressive, Schizophrenia, Anxiety, Bipolar, Eating]
labels = ['Depressive', 'Schizophrenia', 'Anxiety', 'Bipolar', 'Eating']

plt.figure(figsize=(5, 5))
plt.pie(disorders, labels=labels, autopct='%1.1f%%')
plt.title('Average Distribution of Mental Health Disorders')
plt.show()
```

```python
turkey_data = data[data['Entity'] == 'Turkey']

plt.figure(figsize=(8, 5))
plt.plot(turkey_data['Year'], turkey_data['Schizophrenia disorders (share of population) - Sex: Both - Age: Age-standardized'], marker='o')
plt.title('Schizophrenia Disorders (Share of Population) in Turkey')
plt.xlabel('Year')
plt.ylabel('Schizophrenia Disorders (Share of Population) (%)')
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(12, 6))
sns.barplot(x='Year', y='Eating disorders (share of population) - Sex: Both - Age: Age-standardized', data=turkey_data, color='b', label='Eating Disorders')
sns.barplot(x='Year', y='Bipolar disorders (share of population) - Sex: Both - Age: Age-standardized', data=turkey_data, color='g', alpha=0.6, label='Bipolar Disorders')
plt.title('Eating and Bipolar Disorders (Share of Population) in Turkey - Age-standardized')
plt.xlabel('Year')
plt.ylabel('Disorders (Share of Population) (%)')
plt.xticks(rotation=45)
plt.legend()
plt.show()
```

```python
countries = ['China', 'Turkey', 'United States', 'India', 'Brazil']
years = [2000, 2010, 2019]
selected_data = data[(data['Entity'].isin(countries)) & (data['Year'].isin(years))]
selected_data.loc[:, 'Year'] = selected_data['Year'].astype(str)
disorder = 'Anxiety disorders (share of population) - Sex: Both - Age: Age-standardized'
plt.figure(figsize=(12, 4))
sns.barplot(x='Entity', y=disorder, hue='Year', data=selected_data, )
plt.xlabel("Country")
plt.ylabel("Share of Population (%)")
plt.title("Anxiety Disorders in Selected Countries Over the Years")
plt.xticks(rotation=45)
plt.show()
```

```python
years = [1990, 2000, 2010, 2019]
turkey = data[(data['Entity'] == 'Turkey') & (data['Year'].isin(years))]
turkey.loc[:, 'Year'] = turkey['Year'].astype(str)

disorders = [col for col in turkey_data if 'disorders' in col]
melted_turkey_data = turkey.melt(id_vars='Year', value_vars=disorders, var_name='Disorder', value_name='Value')
melted_turkey_data['Disorder'] = melted_turkey_data['Disorder'].str.split().str[0]
melted_turkey_data['Value'] = melted_turkey_data['Value']

plt.figure(figsize=(12, 5))
sns.barplot(x='Disorder', y='Value', hue='Year', data=melted_turkey_data, palette="viridis")
plt.xlabel("Disorder")
plt.ylabel("Share of Population (%)")
plt.title("Mental Health Disorders in Turkey Over 2015-2019")
plt.xticks(rotation=45)
plt.legend(title='Year')
plt.tight_layout()
plt.show()
```
